```python
# map можно использовать для замены значений в колонке
# нужно передать в него словарь типа {'old_val': 'new_val'}

# аналогично работает метод replace 

df_copy = df.copy()
vals = {0:'dead', 1:'survived'}
df_copy['Churn'] = df_copy['Churn'].astype('object')
df_copy['Churn'] = df_copy['Churn'].map(vals)
df_copy['Churn'].value_counts()
```

```python
# Группировка данных
# в общем случае выглядит так
# df.groupby(by=grouping_columns)[columns_to_show].function()

# 1. К датафрейму применяется метод groupby, который разделяет данные по grouping_columns – признаку или набору признаков.
# 2. Индексируем по нужным нам столбцам (columns_to_show).
# 3. К полученным группам применяется функция или несколько функций.
```

```python
cols_to_show = ['Total day minutes', 'Total eve minutes', 'Total night minutes']
df.groupby(by=['Churn'])[cols_to_show].describe()
```

```python
#  можно применить ф-ю agg и передать в нее список ф-ций которые будут применены
df.groupby(by=['Churn'])[cols_to_show].agg([np.mean, np.min, np.max])
```

```python
# Преобразование датафреймов
# Добавление нового столба
```

```python
total_calls = df['Total day calls'] + df['Total eve calls'] + \
              df['Total night calls'] + df['Total intl calls']
df.insert(loc=len(df.columns), column='Total calls', value=total_calls)
```

```python
df.head()
```

```python
# Если создаем из имеющихся столбцов то можно не создавать промежуточный Series
df['Total charge'] = df['Total day charge'] + df['Total eve charge'] + \
                     df['Total night charge'] + df['Total intl charge']
```
